```python
# Create a dictionary mapping EPA Reg Number to Product Name and Units
pesticide_dict = unique_pesticides_dictionary.set_index("EPA Reg Number")[["Product Name", "Units", "Dosage Rate"]].to_dict("index")
```

```python
# Function to expand rows based on EPA Reg Numbers found in "Pest Control Material Used" so each pesticide used on a job has it's own row
def expand_row(row, pesticide_dict):
    # Find all EPA Reg Numbers in the current row's "Pest Control Material Used" column
    epa_numbers = re.findall(r'\d{4,}-\d+', row["Pest Control Material Used"])

    # Find all quantities using regex based on the "Quantity Used (Include unit type please):" pattern
    # The regex will capture everything after "Quantity Used" and stop before "Pest Control Material Type", a comma, or end of the string
    quantities = re.findall(r'Quantity Used \(Include unit type please\):\s*([\w\s/]+?)(?:,|$|Pest)', row["Pest Control Material Used"])

    # Creating a list to store the expanded rows
    expanded_rows = []

    for i, epa in enumerate(epa_numbers):
        if epa in pesticide_dict:
            # Create a new row with the original data plus the matched pesticide info
            new_row = row.copy()
            new_row["EPA Reg Number"] = epa
            new_row["Product Name"] = pesticide_dict[epa]["Product Name"]
            new_row["Units"] = pesticide_dict[epa]["Units"]
            new_row["Dosage Rate"] = pesticide_dict[epa]["Dosage Rate"]

            # Attach the associated quantity if available
            quantity = quantities[i] if i < len(quantities) else ''
            new_row["Quantity Used"] = quantity.strip()  # Ensure it's stripped of leading/trailing spaces

            expanded_rows.append(new_row)

    return expanded_rows
```

```python
# Apply the function to expand the rows
expanded_data = df.apply(lambda row: expand_row(row, pesticide_dict), axis=1)
```

```python
# Flatten the list of lists into a DataFrame
flattened_list = [item for sublist in expanded_data for item in sublist]

# Create the DataFrame
expanded_df = pd.DataFrame(flattened_list)

# Optionally reset the index to ensure uniqueness
expanded_df = expanded_df.reset_index(drop=True)

# Display the DataFrame
print(expanded_df)
```
