### Step 3: Using the slug, scrape each restaurant's page and create a single list of food dictionaries.

Your list of foods should look something like this:
```python
foods = [
    {
        'calories': '0',
        'carbs': '0',
        'category': 'Drinks',
        'fat': '0',
        'name': 'A&W® Diet Root Beer',
        'restaurant': 'A&W Restaurants'
    },
    {
        'calories': '0',
        'carbs': '0',
        'category': 'Drinks',
        'fat': '0',
        'name': 'A&W® Diet Root Beer',
        'restaurant': 'A&W Restaurants'
    },
    ...
]
```

**Note**: Remove extra white space from each category

````python
for i in range(1, 105):
    url = 'http://chubbygrub.com/restaurants/' + str(i)
    
    restaurants = soup.find_all('a', {'class': 'btn btn-lg btn-primary'})
    restaurants_cleaned = [i.text for i in restaurants]
    restaurant_links = [i.attrs['href'] for i in restaurants]
    restaurant_slugs = [i['href'].replace('/restaurants/', '') for i in restaurants]
    restaurant_list = [{"restaurants": n, "slug": i} for n,i in zip(restaurants_cleaned, restaurant_slugs)]
````

````python
foods = []
for name_and_slug in restaurant_list:    
    url = 'http://chubbygrub.com/restaurants/' + name_and_slug["slug"]
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    nutrition_info = soup.find_all('tr', {'itemtype': "http://schema.org/NutritionInformation"})
    for n in nutrition_info: 
        food_dict = {
            'calories': n.find('td', {'itemprop': 'calories'}).text,
            'carbs': n.find('td', {'itemprop': 'carbohydrateContent'}).text,
             'category':n.find('td', {'class': 'hidden-xs category-column'}).find('a').text,
             'fat':n.find('td', {'itemprop': 'fatContent'}).text,
             'name': n.find('td', {'itemprop': 'name'}).text,
             'restaurant': name_and_slug["restaurants"]}
        foods.append(food_dict)
                                         
                
foods
         
        
````
